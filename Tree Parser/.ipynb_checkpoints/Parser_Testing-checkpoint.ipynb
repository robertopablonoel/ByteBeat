```python
def selectOp(tree, max_height = 2, height = None, node = None):
    tree_ops = dict(tree.operators())
    if height == None:
        low_ops = [i for i in tree_ops if tree_ops[i] <= max_height]
    else:
        low_ops = [i for i in tree_ops if tree_ops[i] == height]
    height = max([tree_ops[i] for i in low_ops])
    inverted = [height+1-tree_ops[i] for i in low_ops]
    total = sum(inverted)

    nodeProbs = [i / total for i in inverted]
    nodeVals = low_ops
    
    nodeChoice = random.choice(nodeVals, p=nodeProbs)
    print("Operator %s at height %d" % (nodeChoice._element, tree_ops[nodeChoice]))
    return (nodeChoice, height)
```
