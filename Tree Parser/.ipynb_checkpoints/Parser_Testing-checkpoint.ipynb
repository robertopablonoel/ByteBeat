```python
parent = "((t>>4)|(t%11))|(((t%102)|(t>>9))&((t>>9)|((t*t)%16)))"
tree = makeTree(parent)
```

```python
#pretty_print(tree)
```

```python
parentParsed = prettyParse(tree)
print(parentParsed == parent)
print(parentParsed + "\n" + parent)
```

```python
print(parentParsed)
pretty_print(tree)
```

```python
fs = 10000
seconds = 30

#must %256 so it doesn't go out of char range
wave1 = [eval(parentParsed) % 256 for t in range(fs * seconds)]

plt.plot(wave1,'.')
plt.show()

arrOG = np.array(wave1)
f, t, Sxx = signal.spectrogram(arrOG, 16000)
plt.pcolormesh(t, f, Sxx)
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()

IPython.display.Audio(data=wave1, rate=fs)
```
