```python
def randomTree():
    tree = Tree()
    tree.add_root(createTree(spaceOut(expression)))
    max_nodes = 20
    tree._root = randomTreeSupport(max_nodes, 0)
    return tree

def randomTreeSupport():
    oVals = ['>>','*','&','^','|','-','+','/']
    oProbs = [.2,.2,.2,.1 + 2/30,.1,0,2/30,2/30]
    isleaf = random.choice([0,1], p=[.5,.5])
    
    if isleaf:
        root = int(np.round(np.random.normal(loc = 30, scale=20)))
        return root
    else:
        root = random.choice(oVals, p=oProbs)
        root._left = randomTreeSupport()
        root._right = randomTreeSupport()
        return root
        
    
```
